```python
import os
import pandas as pd
import numpy as np
import librosa
from tqdm import tqdm
```

```python
# Define file paths
metadata_path = 'fma_metadata/tracks.csv' 
audio_dir = 'dataset'
```

```python
tracks = pd.read_csv(metadata_path, header=[0,1], index_col=0)
```

```python
audio_files = []
for root, dirs, files in os.walk(audio_dir):
    for file in files:
        if file.endswith('.mp3'):
            audio_files.append(os.path.join(root, file))
```

```python
# Create an empty DataFrame to store features
features_df = pd.DataFrame(columns=['track_id', 'duration', 'mfcc'])
```

```python
features_df
```

```python
# Function to extract MFCC features
def extract_mfcc_features(file_path, sr=22050):
    try:
        # Load audio file
        audio, sr = librosa.load(file_path, sr=sr)
        
        # Extract MFCC features
        mfcc = librosa.feature.mfcc(y=audio, sr=sr, n_mfcc=13)
        
        # Calculate duration
        duration = librosa.get_duration(y=audio, sr=sr)
        
        # Calculate mean of MFCC features
        mfcc_mean = np.mean(mfcc, axis=1)
        
        return duration, mfcc_mean.tolist()
    
    except (librosa.util.exceptions.ParameterError, Exception) as e:
        print(f"Error processing file {file_path}: {e}. Skipping this file.")
        return None, None
```

```python
# Iterate over audio files with tqdm
for file_path in tqdm(audio_files[:1500], desc="Processing audio files"):
    track_id = int(os.path.basename(file_path)[:-4])
    duration, mfcc_features = extract_mfcc_features(file_path, sr=22050)
    
    if duration is not None and mfcc_features is not None:
        features_df.loc[len(features_df)] = {'track_id': track_id, 'duration': duration, 'mfcc': mfcc_features}
```

```python
# Round duration and convert track_id to int
features_df['track_id'] = features_df['track_id'].astype(int)
features_df['duration'] = features_df['duration'].round(2)

```

```python
# Print summary of features DataFrame
print("Summary of features:")
print(features_df)
```
